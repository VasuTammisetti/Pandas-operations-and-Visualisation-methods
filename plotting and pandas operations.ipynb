```python
def match_words(words):
    ctr = 0
    
    
    for word in words:
        if len(word) > 1 and word[0] == word[-1]:
            ctr += 1
    return ctr
```

```python
a=['abc', 'xyz', 'aba', '1221','bhgsskknb','aa']
```

```python
match_words(a)
```
